### III) Slicing Rows & Columns with .loc

```python
df.loc["muffin",'Price']
```

Slicing 1 row and multiple columns at the same time passing an index ("row") and a list of specific labels ("columns") -> return a Series

```python
df.loc["muffin",['Price','Student']]
```

Slicing multiple rows and columns at the same time passing a list of indexes ("rows") and labels ("columns") -> return a Series -> returns a DataFrame

```python
df.loc[ ['muffin','scone','banana bread'], ['Price','Expense','Student']]
```

Slicing multiple rows and columns, savings the lists of rows and column indices as variables first.

```python
# Save lists of rows
rows_to_slice =  ['muffin','scone','banana bread']
# Save list of columns
cols_to_slice =  ['Price','Expense','Student']
# Use variables of lists to slice
df.loc[rows_to_slice, cols_to_slice]
```
