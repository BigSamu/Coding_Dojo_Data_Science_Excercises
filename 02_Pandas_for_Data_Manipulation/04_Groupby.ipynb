## Question and Answers Core Excercise

1) What is the average age for the entire data set?

```python
# Method 1 -> using [] brackets
df_insurance["age"].mean()
```

```python
# Method 2 -> using .loc()
df_insurance.loc[:,"age"].mean()
```

2) What is the age of the oldest person in the entire data set (the maximum age)?

```python
df_insurance["age"].max()
```

3) What is the age of the youngest person in the entire data set (the minimum age)?

```python
df_insurance["age"].min()
```

4) What is the average for charges for the entire data set?

```python
df_insurance["charges"].mean()
```

5) What is the average for charges grouped by sex?

```python
mean_charges_by_sex = df_insurance.groupby("sex")["charges"].mean()
mean_charges_by_sex
```

6) What is the average for charges, grouped by if the person is a smoker?

```python
mean_charges_if_smoker = df_insurance.groupby("smoker")["charges"].mean()
mean_charges_if_smoker
```

7) What is the average for charges grouped by if the person smokes and their sex?

```python
mean_charges_if_smoker_and_sex = df_insurance.groupby(["smoker","sex"])["charges"].mean()
mean_charges_if_smoker_and_sex
```
